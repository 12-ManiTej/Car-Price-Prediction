```python
from sklearn.preprocessing import LabelEncoder
```

```python
data.drop(['car_ID','symboling','CarName'],axis=1,inplace=True)
```

```python
l=LabelEncoder()
```

```python
data["fueltype"]=l.fit_transform(data['fueltype'])
data["aspiration"]=l.fit_transform(data['aspiration'])
data["doornumber"]=l.fit_transform(data['doornumber'])
data["carbody"]=l.fit_transform(data['carbody'])
data["drivewheel"]=l.fit_transform(data['drivewheel'])
data["enginelocation"]=l.fit_transform(data['enginelocation'])
data["fuelsystem"]=l.fit_transform(data['fuelsystem'])
data["enginetype"]=l.fit_transform(data['enginetype'])
data["cylindernumber"]=l.fit_transform(data['cylindernumber'])
```

```python
data
```

```python
data=data[['drivewheel','carlength','carwidth','curbweight','enginesize','horsepower','citympg','highwaympg','price']]
```

```python
data
```

```python
X=data[['drivewheel','carlength','carwidth','curbweight','enginesize','horsepower','citympg','highwaympg']]
y=data['price']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=0)
```

```python
print(X_train.shape,X_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model=LinearRegression()
model.fit(X_train,y_train)
```

```python
plt.scatter(y_train, model.predict(X_train), label='Training Data', color='blue')
```

```python
plt.scatter(y_test, model.predict(X_test), label='Testing Data', color='orange')

```

```python
plt.plot([min(y_train), max(y_train)], [min(model.predict(X_train)), max(model.predict(X_train))], linestyle='--', color='red', label='Best Fit Line')
plt.xlabel('Actual Price')
plt.ylabel('Predicted Price')
plt.title('Car Price Prediction')
plt.legend()
plt.show()
```

```python
y_pred=model.predict(X_test)
```

```python
from sklearn.metrics import r2_score
print("r2 score=%s" %r2_score(y_test,y_pred))
```
